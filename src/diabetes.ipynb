```python
import pandas as pd

# Step 1: Load the dataset
df_diabetes = pd.read_csv("../data/diabetes_unclean.csv")

# Display first few rows to understand its structure
print("Initial data preview:")
print(df_diabetes.head())
```

```python
# Step 3: Display basic statistics
print("\nBasic statistics:")
print(df_diabetes.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of each gender
gender_counts = df_diabetes['Gender'].value_counts()

# Plotting bar graph
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting line graph
plt.figure(figsize=(10, 6))
sns.histplot(df_diabetes['AGE'], kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Count of each class
class_counts = df_diabetes['CLASS'].value_counts()

# Plotting pie chart
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Classes')
plt.show()
```

```python
# Plotting histogram
plt.figure(figsize=(10, 6))
sns.histplot(df_diabetes['Urea'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Urea')
plt.xlabel('Urea')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
