```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv("../data/winequality-red.csv")

# Step 2: Display the first few rows of the dataframe to understand its structure
print("Initial data preview:")
print(df.head())

# Step 3: Check the shape of the dataset
print("\nShape of the dataset:")
print(df.shape)

# Step 4: Check for any missing values in the dataset
print("\nChecking for missing values:")
print(df.isnull().sum())

# Step 5: Display basic statistics of numeric columns
print("\nBasic statistics:")
print(df.describe())

# Step 6: Visualize the distribution of quality ratings using a bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=df, palette='viridis')
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality Rating')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Step 7: Visualize the correlation between numeric variables using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Step 8: Visualize the distribution of alcohol content using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['alcohol'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
