```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv("../data/Life Expectancy Data.csv")

# Step 2: Display the first few rows of the dataframe to understand its structure
print("Initial data preview:")
print(df.head())

# Step 3: Check for any missing values in the dataset
print("\nChecking for missing values:")
print(df.isnull().sum())

# Step 4: Clean the dataset by filling missing values with the mean of each column
df.fillna(df.mean(), inplace=True)

# Step 5: Verify that no missing values remain after cleaning
print("\nVerifying no missing values after cleaning:")
print(df.isnull().sum())

# Step 6: Display basic statistics after cleaning
print("\nBasic statistics after cleaning:")
print(df.describe())

# Step 7: Visualization - Example plots

# Example 1: Histogram of Life Expectancy
plt.figure(figsize=(8, 6))
sns.histplot(df['Life_expectancy'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Example 2: Scatter plot of GDP vs Life Expectancy
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GDP', y='Life_expectancy', data=df, color='coral')
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()

# Step 8: Save the cleaned dataset to a new CSV file
df.to_csv("cleaned_life_expectancy.csv", index=False)

print("\nCleaned data saved to 'cleaned_life_expectancy.csv'.")
```

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("../data/Life Expectancy Data.csv")

# Step 2: Display the initial shape of the dataset
print("Initial shape of the dataset:", df.shape)

# Step 3: Display the first few rows of the dataframe to understand its structure
print("\nInitial data preview:")
print(df.head())

# Step 4: Check for any missing values in the dataset
print("\nChecking for missing values:")
print(df.isnull().sum())

# Step 5: Clean the dataset by filling missing values with the mean of each numeric column
numeric_cols = df.select_dtypes(include=['number']).columns  # Select numeric columns only
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())  # Fill NaN with mean of each numeric column

# Step 6: Verify that no missing values remain after cleaning
print("\nVerifying no missing values after cleaning:")
print(df.isnull().sum())

# Step 7: Display basic statistics after cleaning
print("\nBasic statistics after cleaning:")
print(df.describe())

# Step 8: Display the final shape of the dataset after cleaning
print("\nFinal shape of the dataset:", df.shape)

# Step 9: Save the cleaned dataset to a new CSV file
df.to_csv("cleaned_life_expectancy.csv", index=False)

print("\nCleaned data saved to 'cleaned_life_expectancy.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the cleaned dataset
df = pd.read_csv("cleaned_life_expectancy.csv")

# Step 2: Visualize life expectancy distribution using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Life expectancy '], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Step 3: Visualize life expectancy trends over years using a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Life expectancy ', data=df, ci=None)
plt.title('Life Expectancy Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()

# Step 4: Visualize a bar graph of status counts
plt.figure(figsize=(8, 6))
sns.countplot(x='Status', data=df, palette='viridis')
plt.title('Distribution of Countries by Development Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Step 5: Visualize a pie chart of status proportions
status_counts = df['Status'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Countries by Development Status')
plt.axis('equal')
plt.show()
```
